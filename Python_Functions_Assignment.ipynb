***Que 5:- What are iterators in Python and how do they differ from iterables?***

Ans:- An iterator is an object in Python that represents a stream of data and returns the next value using the __next__() method. It keeps track of its current position and raises a StopIteration exception when no more elements are left. An iterable, on the other hand, is any object that can be looped over (like lists, strings, tuples) and implements the __iter__() method, which returns an iterator. While all iterators are iterables, not all iterables are iterators. To make an iterable behave like an iterator, it needs to be passed to the iter() function.

**Que 6:- Explain the concept of generators in Python and how they are defined?**

Ans:- Generators in Python are a type of iterable that generate values and do not store the entire sequence in memory. They are defined using functions but instead of the return keyword, they use the yield keyword to return values one at a time. Each time the generator function is called, it resumes from where it last yielded a value. This makes generators memory-efficient, especially when dealing with large datasets. Example:

def count_up_to(n):

    i = 1
    while i <= n:
        yield i
        i += 1




**Que 7:- What are the advantages of using generators over regular functions?**

Ans:-Generators offer various advantages over functions:

    --> Generators yield items one at a time and do not store the entire result in memory, unlike functions that return full lists so they are memory efficient.

    --> Since values are generated on demand, the initial response is faster.

    --> Generators can produce an infinite stream of data without memory overflow.

    --> They allow writing clean and concise code using the yield statement, eliminating the need for manually managing things.



**Que 8:- What is a lambda function in Python and when is it typically used?**

**Que 9 :-  Explain the purpose and usage of the map() function in Python?**

**Que 10:- What is the difference between map(), reduce(), and filter() functions in Python?**

Ans:- map(): Applies a function to each item in an iterable and returns a new iterator with the transformed items.

filter(): Applies a function that returns a boolean to each item and filters out those for which the function returns False.

reduce() : Applies a function to the items in an iterable, reducing it to a single value.

```
from functools import reduce

numbers = [1, 2, 3, 4]

print(list(map(lambda x: x*2, numbers)))  
Output --> [2, 4, 6, 8]

print(list(filter(lambda x: x%2 == 0, numbers)))
Output --> [2, 4]

print(reduce(lambda x, y: x + y, numbers))   
Output --> 10
```

````python
# Que 1:-  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list

def sum_even(num_list):
  even_sum = 0
  for num in num_list:
    if num % 2 == 0:
      even_sum += num
  return even_sum

my_list = [10, 15, 22, 33, 40]
result = sum_even(my_list)
print("Sum of even numbers are ", result)
````

````python
# Que 2:- Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(string_rev):
  string1 = string_rev[::-1]
  return string1

string2 = "I am Utkarsh"
result = reverse_string(string2)
print("The reversed string is -> ", result)
````

````python
# Que 3:- . Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_of_num(int_list):
  square_list = []
  for x in int_list:
    square_list.append(x**2)
  return square_list

int_list = [1,2,3,4,5]
result = square_of_num(int_list)

print("Squared list is --> ", result)
````

````python
# Que 4:- Write a Python function that checks if a given number is prime or not from 1 to 200.

def check_prime(num):
  if num < 2:
    return False
  for i in range(2 , int(num**0.5) + 1):
    if num % i == 0:
            return False
  return True

for num in range(1, 201):
    if check_prime(num):
        print(f"{num} is a prime number")
````

````python
# Que 5:-Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fibonacci(n):
    sequence = []
    a, b = 0, 1

    if n <= 0:
        return sequence

    for _ in range(n):
        sequence.append(a)
        a, b = b, a + b

    return sequence

print(fibonacci(10))
````

````python
# Que 6- Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

for value in powers_of_two(5):
    print(value)
````

````python
# Que 7:-  Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')


for line in read_file_line_by_line("/content/Llama3 model progress.txt"):
    print(line)
````

````python
# Que 8:- Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

data = [(4, 10), (1, 5), (3, 2), (2, 8)]

sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
````

````python
# Que 9:-Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celsius = [0, 10, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Temperatures in Fahrenheit:", fahrenheit)
````

````python
# Que 10:-Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'

    result = ''.join(filter(lambda char: char not in vowels, input_str))
    return result

text = "Hello, I am Utkarsh"
no_vowels = remove_vowels(text)
print("String without vowels:", no_vowels)
````
